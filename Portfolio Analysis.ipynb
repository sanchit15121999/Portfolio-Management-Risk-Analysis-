```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
```

```python
#this is a guy named as Sd and he invests in only tech giants
SD= ['MSFT','NFLX','FB','AMZN']
ND=['MSFT','PFE','F','WMT']


df_tech=pd.DataFrame()
df_div=pd.DataFrame()

#taking data from web quandl could have also been taken from google finance or yahoo
for tech,div in zip(SD,ND):
    df_tech[tech]=web.DataReader(tech, data_source='yahoo', start='2017-1-1')['Adj Close']
    df_div[div]=web.DataReader(div, data_source='yahoo', start='2017-1-1')['Adj Close']
```

## Risk is lesser in div as compared to techie . But the reward is also less

### Compunded daily return

```python
#250 is th number of trading days and here we have the annual return of stocks
avg_return_tech=(pow(df_tech.iloc[850]/df_tech.iloc[0],1/850)-1)*851
avg_return_div=(pow(df_div.iloc[850]/df_div.iloc[0],1/850)-1)*851
```

```python
#these are the avg returns from  ie 125% for netflix
print(avg_return_tech)
print(avg_return_div)
```

```python
#to calculate the return on the portfolio that is weighted average returns on these stocks
#we assume tht each of the stocks have beeen equally invested in
#to we will multiply the returns by 
#we will get a matrix in return
weights=np.array([0.25,0.25,0.25,0.25])
tech_portfolio_return=np.dot(weights,avg_return_tech)
div_portfolio_return=np.dot(weights,avg_return_div)
#we will get a single number
print(tech_portfolio_return)
print(div_portfolio_return)
```

```python
#now to understand the risk of the portflio
#we will use the concept of covarience 
#we are calculating daily returns/returns of the previous days prices -1 
#daily returns of both days 
#shift 1 will shift the division column to 1 down
tech_daily=((df_tech/df_tech.shift(-1))-1)
div_daily=((df_div/df_div.shift(-1))-1)
```

# Portfolio Risk

```python
#risk of portfolio for 2 stocks is given by w1^2sigma1square+  + 2w1w2cov(1,2)
covr_tech = tech_daily.cov()*250
covr_div = div_daily.cov()*250
#we will get a covarience matrix and the diagonal represents covarience wtih itsef and the rest will show covarience with others
print(covr_tech)
print(covr_div)
```

```python
#we will use matrix multiplication formula  to get the varience 
var_te= np.dot(weights.T,np.dot(covr_tech,weights))
var_div= np.dot(weights.T,np.dot(covr_div,weights))
```

```python
#risk is the standard deviation or the under root of varience
tech_risk=var_te**0.5
div_risk=var_div**0.5
```

```python
print(tech_risk)
print(div_risk)
```

# Sharp raito

```python
rf=0.02
#usinhg this we will calculate which stock is givinh you higher reward for the risk that they are taking
sharp_ratio_te=(tech_portfolio_return-rf)/tech_risk
sharp_ratio_div=(div_portfolio_return-rf)/div_risk
```

```python
#so for ____ risk tech is giving this much higher reward
print(sharp_ratio_te)
print(sharp_ratio_div)
```
